```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./data/weight-height.csv")
df.info()
df.head(2)
```

```python
x = df[['Height']].values
y_true = df[['Weight']].values
```

```python
# Line Equation Function
def line(x, w = 0, b = 0):
    return w * x + b

# Loss Function
def mean_square_error(y_true, y_pred):
    s = (y_true - y_pred)**2
    return np.sqrt(s.mean())
```

```python
plt.figure(figsize = (10, 5))
ax1 = plt.subplot(121)
df.plot(kind = 'scatter',
       x = 'Height',
       y = 'Weight',
       title = 'Weight and Height in Adults', ax = ax1)

bs = np.array([-100, -50, 0, 50, 100, 150])
mses = []

for b in bs:
    y_pred = line(x, w = 2, b = b) # Adjust w and b for find best
    mse = mean_square_error(y_true, y_pred)
    mses.append(mse)
    plt.plot(x, y_pred)
    
ax2 = plt.subplot(122)
plt.plot(bs, mses, 'o-')
plt.title('Cost Function of b')
plt.xlabel('b')
```

```python
plt.figure(figsize = (10, 5))
ax1 = plt.subplot(111)
df.plot(kind = 'scatter',
       x = 'Height',
       y = 'Weight',
       title = 'Weight and Height in Adults', ax = ax1)

y_pred = line(x, w = 7.715593, b = -348.13806) # Adjust w and b for find
    

plt.plot(x, y_pred, color = 'red')
plt.title('Cost Function of b')
```

# Find W and B value with Deep Learning

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam, SGD

%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
print(x[:2])
print(y_pred[:2])
```

```python
x.shape
```

```python
model = Sequential()

# model structure
model.add(Dense(1, input_shape = (1,)))
```

```python
model.summary()
```

```python
model.compile(Adam(lr = 0.8), "mean_squared_error")
```

```python
model.fit(x, y_true, epochs = 40)
```

```python
y_pred = model.predict(x)
y_pred
```

```python
df.plot(kind = "scatter", 
       x = "Height", 
       y = "Weight", 
       title = "Height vs Weight with Deep Learning")

plt.plot(x, y_pred, color = 'red', linewidth = 3)
```

```python
w, b = model.get_weights()
print(w, b, sep = "\n\n")
```
