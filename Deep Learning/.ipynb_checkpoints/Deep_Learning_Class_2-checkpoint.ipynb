```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(np.random.randint(10, 100, size = (20000, 5)), 
                 columns = ["s1", "s2", "s3", "s4", "s5"])
df.head()
```

```python
# First way
df['obtained'] = df.loc[:,'s1':'s5'].apply(lambda x:np.sum(x), axis = 1)
df.head()
```

```python
# Second way
df['obtained'] = df.s1 + df.s2 + df.s3 + df.s4+ df.s5
df.head()
```

```python
# Third way
df['obtained'] = df.loc[:,'s1':'s5'].apply(sum, axis = 1)
df.head()
```

```python
# Fourth way
def cal_sum(*x):
    return np.sum(x)
df['obtained'] = df.loc[:,'s1':'s5'].apply(lambda x:np.sum(x), axis = 1)
df.head()
```

```python
df['Total'] = 500
df.head()
```

```python
df['percentage'] = df.obtained / df.Total * 100
df.head()
```

# Calculate Grade With Rule Based Programming
### Symbolic A.Ic(Classical Programming)
### Traditional Programming

```python
# Define Rule (Always Defined By Human(Programmer))

def grade(x):
    if x >= 80:
        return "A+"
    elif x >= 70:
        return "A"
    elif x >= 60:
        return "B"
    elif x >= 50:
        return "C"
    elif x >= 40:
        return "D"
    elif x >= 33:
        return "A"
    else:
        return "Fail"
    
df['Grade'] = df.percentage.apply(grade)
df.head()
```

```python
df.Grade.value_counts()
```

```python
df.corr().plot(kind = 'bar')
```

```python
df.Grade.unique()
```

```python
# This way cant work properly
dic = {v:k for k,v in enumerate(df.Grade.unique())}
dic
```

```python
dic = {
    "A+" : 7,
    "A" : 6, 
    "B" : 5,
    "C" : 4,
    "D" : 3,
    "E" : 2,
    "Fail" : 1
}
dic
```

```python
df['Grade1'] = df.Grade.map(dic)
df.head()
```

```python
df.corr().plot(kind = 'bar')
```

```python
df1 = df[['s1', 's2', 's3', 's4', 's5', 'Grade']]
df1.head()
```

# Apply Deep Learning to get rules or Model or Software

```python
df1.head(2)
```

### Apply one hot  - encode on Grade column

```python
pd.get_dummies(df1.Grade).head()
```

```python
x = df1.loc[:, 's1':'s5']
x.head()
```

```python
y = df1.Grade
y = pd.get_dummies(y) # apply one-hot-encoding on Grade
y.head()
```

# Import Modules

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam, SGD
```

```python
# first-layer input_shape (no of columns in x)
# last-layer output no of labels in Y(unique values)
# last-layer activation function depend on your problem(...)

model = Sequential()
model.add(Dense(5, activation='relu',input_shape=(5,)))
model.add(Dense(6, activation='softmax')) # last - layer output depends on numbers of labels
```

```python
model.summary()
```

# Compile

```python
# Optimizer update -> weights (w,b) based on loss function
model.compile(optimizer=SGD(lr=0.8),
             loss='categorical_crossentropy',
             metrics=['acc'])
```

```python
model.fit(x,y, epochs=30)
```

```python
df['Predict'] = y.columns.take(np.argmax(
    model.predict(df1.loc[:,'s1':'s5']),
    axis=1
))
df.head()
```
